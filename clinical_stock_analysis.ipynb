```python
# final thing....create column to show short and long expiry dates
import datetime
today = datetime.datetime.now()

# 1. Calculate days remaining
# We use .dt.days to get a simple number of days
stock_clean['days_to_expiry'] = (stock_clean['expiry_date'] - today).dt.days

# 2. Categorize them
def categorize_expiry(days):
    if days < 0:
        return 'Expired'
    elif days <= 90:
        return 'Short Expiry (<3 Months)'
    elif days <= 180:
        return 'Medium Expiry (3-6 Months)'
    else:
        return 'Long Expiry (>6 Months)'

stock_clean['expiry_status'] = stock_clean['days_to_expiry'].apply(categorize_expiry)
```

```python
# Step 1: Define your ACTUAL dead stock list (from the green square)
true_dead_stock_items = [
    'synthecon 5', 'synthecon 6', 'polyproplene', 
    'surgicril 6/0', 'advacryl 6/0', 'dermatech 4', 'dermacryl 5'
]

# Step 2: Define critical surgical supplies (keep even if zero movement)
critical_surgical_supplies = [
    'vicryl sutures', 'silk sutures', 'nylon sutures', 'endotracheal tube',
    'catheter', 'folleys', 'chest drainage', 'feeding tube', 'sunction catheter',
    'tracheostomy', 'cannula', 'gauze', 'speculum', 'drainage bottle'
]

# Step 3: Define equipment (not consumed)
equipment_items = [
    'scale', 'stethoscope', 'machine', 'bp', 'sphygmomanometer', 
    'bathroomscale', 'thermometer', 'stethoscope'
]

def identify_stock_type(row):
    item_name = str(row['item_name']).lower()
    
    # RULE 1: MANUAL OVERRIDE - These are the ONLY true dead stock
    if any(dead_item in item_name for dead_item in true_dead_stock_items):
        return 'Dead Stock (Confirmed - Discontinue)'
    
    # RULE 2: EQUIPMENT - Not consumable stock
    if any(equip in item_name for equip in equipment_items):
        return 'Equipment (Not Consumable)'
    
    # RULE 3: CRITICAL SURGICAL SUPPLIES - Essential even with zero movement
    if any(critical in item_name for critical in critical_surgical_supplies):
        if row['variance'] == 0 and row['opening_stock'] > 0:
            return 'Critical Safety Stock (Essential)'
        elif row['variance'] < -10:
            return 'Fast Mover (Surgical)'
        else:
            return 'Normal Movement (Surgical)'
    
    # RULE 4: STOCKOUT (Both opening and closing are zero)
    if row['opening_stock'] == 0 and row['closing_stock'] == 0:
        return 'Stockout (Critical)'
    
    # RULE 5: FAST MOVER (High consumption)
    elif row['variance'] < -10:
        return 'Fast Mover'
    
    # RULE 6: NORMAL MOVEMENT (Some consumption)
    elif row['variance'] != 0:
        return 'Normal Movement'
    
    # RULE 7: ZERO MOVEMENT BUT NOT DEAD (e.g., safety stock for non-critical items)
    elif row['variance'] == 0 and row['opening_stock'] > 0:
        return 'Safety Stock (No Recent Use)'
    
    # RULE 8: CATCH-ALL
    else:
        return 'Review Manually'
```

```python
def final_clean_drill(row):
    # 1. PROTECT THE STOCKOUTS (Priority #1)
    # If both are zero, it's a Stockout, no matter what it is.
    if row['opening_stock'] == 0 and row['closing_stock'] == 0:
        return 'Stockout (Critical)'
    
    # 2. HANDLE THE THEATRE ESSENTIALS (The "Static" ones)
    item_name = str(row['item_name']).lower()
    theatre_keywords = ['sutures', 'catheter', 'tube', 'speculum', 'gauze', 'tape', 'machine', 'mva']
    
    if any(word in item_name for word in theatre_keywords) and row['variance'] == 0:
        return 'Active Stock (Static Count)'
    
    # 3. HANDLE ACTUAL DEAD STOCK
    # NOTE: This is where you'll put your manual dead stock list
    true_dead_stock = ['synthecon', 'polyproplene', 'surgicril', 'advacryl', 'dermatech', 'dermacryl']
    if any(dead in item_name for dead in true_dead_stock):
        return 'Dead Stock (Frozen Money)'
    
    # 4. HANDLE OTHER ZERO VARIANCE ITEMS (These are NOT dead)
    if row['variance'] == 0 and row['opening_stock'] > 0:
        # Check if it's equipment
        durable_equipment = ['scale', 'stethoscope', 'machine', 'bp', 'sphygmomanometer', 'bathroomscale']
        if any(equip in item_name for equip in durable_equipment):
            return 'Fixed Asset (Not for Consumption)'
        else:
            return 'Safety Stock (No Recent Use)'
    
    # 5. HANDLE FAST MOVERS
    if row['variance'] < -10:
        return 'Fast Mover'
    
    # 6. HANDLE NORMAL MOVEMENT
    if row['variance'] != 0:
        return 'Normal Movement'
    
    # 7. CATCH-ALL (should not reach here)
    return 'Review Manually'

# Apply the new strict logic
stock_clean['stock_analysis'] = stock_clean.apply(final_clean_drill, axis=1)

# VERIFY: This should show your categories
print(stock_clean['stock_analysis'].value_counts())
```

```python
def debug_function(row):
    item_name = str(row['item_name']).lower()
    variance = row['variance']
    opening = row['opening_stock']
    closing = row['closing_stock']
    
    print(f"\n=== Checking: {item_name} ===")
    print(f"Variance: {variance}, Opening: {opening}, Closing: {closing}")
    
    # 1. STOCKOUT
    if opening == 0 and closing == 0:
        print("→ Rule 1: Stockout")
        return 'Stockout (Critical)'
    
    # 2. EQUIPMENT
    durable_equipment = ['scale', 'stethoscope', 'machine', 'bp', 'sphygmomanometer', 'bathroomscale']
    if any(equip in item_name for equip in durable_equipment):
        print("→ Rule 2: Equipment")
        return 'Fixed Asset (Not for Consumption)'
    
    # 3. TRUE DEAD STOCK
    true_dead_stock = ['synthecon 4/0 rb 26 mm']
    if any(dead in item_name for dead in true_dead_stock):
        print("→ Rule 3: True Dead Stock")
        return 'Dead Stock (Frozen Money)'
    
    # 4. NEGATIVE VARIANCE
    if variance < 0:
        print("→ Rule 4: Negative Variance")
        if variance < -10:
            print("  → Fast Mover")
            return 'Fast Mover'
        else:
            print("  → Normal Movement")
            return 'Normal Movement'
    
    # 5. THEATRE ESSENTIALS
    theatre_keywords = ['sutures', 'catheter', 'tube', 'speculum', 'gauze', 'tape', 'mva']
    if any(word in item_name for word in theatre_keywords) and variance == 0:
        print("→ Rule 5: Theatre Essentials")
        return 'Critical Safety Stock (Essential)'
    
    # 6. ZERO VARIANCE
    if variance == 0 and opening > 0:
        print("→ Rule 6: Zero Variance")
        return 'Safety Stock (No Recent Use)'
    
    print("→ Rule 7: Catch-all")
    return 'Review Manually'

# Test on a problematic item
test_row = stock_clean[stock_clean['item_name'].str.contains('Synthecon 5/0 RC 19 MM', case=False, na=False)].iloc[0]
result = debug_function(test_row)
print(f"\nFINAL RESULT: {result}")
```

```python
# CELL A: Define the unit conversion function
# This converts all units like pairs pkts boxes etc into single measures, 50,100,1000 etc

import re  # We need this for finding numbers in text

def get_base_units(unit_string):
    """
    Tells us how many individual items are in 1 unit.
    
    Examples:
    - 'pkts of 100' -> 100 (1 packet = 100 individual items)
    - 'pairs' -> 1 (1 pair = 1 pair - this IS the base unit)
    - 'pks of 50' -> 50
    - 'dozens' -> 1 (for sutures, 1 dozen is the base unit we track)
    - 'pcs' -> 1 (1 piece = 1 piece)
    """
    
    # If we don't know the unit, assume 1
    if pd.isna(unit_string):
        return 1
    
    # Convert to lowercase for easier matching
    unit = str(unit_string).lower()
    
    # Check if this is a "pack of something" (contains 'pk' or 'pkt')
    if 'pk' in unit or 'pkt' in unit:
        # Find any numbers in the text (like '100', '50', '10')
        numbers = re.findall(r'\d+', unit)
        if numbers:
            # Return the first number we find
            return int(numbers[0])
    
    # For everything else (pairs, rolls, pcs, dozens), treat as 1 base unit
    return 1

# Let's test it works
print("Testing the function:")
print(f"  'pkts of 100' -> {get_base_units('pkts of 100')}")  # Should print 100
print(f"  'pairs' -> {get_base_units('pairs')}")              # Should print 1
print(f"  'pks of 50' -> {get_base_units('pks of 50')}")      # Should print 50
print(f"  'dozens' -> {get_base_units('dozens')}")            # Should print 1
```

```python
# STEP 4: Convert stock data to base units
# Run this AFTER STEP 3

# Use unit_x (from stock take) for conversion
stock_final['base_conversion_stock'] = stock_final['unit_x'].apply(get_base_units)

# Create base units column for stock
stock_final['closing_stock_base'] = stock_final['closing_stock'] * stock_final['base_conversion_stock']

# Verify it worked - check the gloves
print("=== STOCK CONVERSION CHECK ===")
print()
gloves_check = stock_final[stock_final['item_name'].str.contains('sterile surgical gloves')]
print(gloves_check[['item_name', 'closing_stock', 'unit_x', 'base_conversion_stock', 'closing_stock_base']].to_string())
```

```python
# STEP 6: Calculate correct Months of Supply
# Run this NOW - using your existing column names

def calculate_correct_mos(row):
    """
    Now both numbers are in the SAME base units!
    """
    stock = row['closing_stock_base']      # e.g., 700 pairs
    consumption = row['consumption_base']  # e.g., 20,000 pairs per month
    
    if pd.isna(consumption) or consumption == 0:
        return None
    
    return round(stock / consumption, 1)

# Create the NEW correct MOS column (not overwriting anything)
stock_final['mos_correct'] = stock_final.apply(calculate_correct_mos, axis=1)

# Show the dramatic difference for problem items
print("=== STOCK vs CONSUMPTION - THE TRUTH REVEALED ===")
print()

# Find items that had mismatched units (where conversion factor > 1)
problem_items = stock_final[
    (stock_final['base_conversion_cons'] > 1) & 
    (stock_final['consumption_base'].notna())
].head(10)

for idx, row in problem_items.iterrows():
    print(f"Item: {row['item_name']}")
    print(f"  Stock: {row['closing_stock']} {row['unit_x']}")
    print(f"  = {row['closing_stock_base']} base units")
    print(f"  Consumption: {row['consumption per month']} {row['unit_y']}")
    print(f"  = {row['consumption_base']} base units per month")
    print()
    
    # Calculate MOS manually so we can show it even if old 'mos' doesn't exist
    if row['consumption_base'] > 0:
        correct_mos = round(row['closing_stock_base'] / row['consumption_base'], 1)
        print(f"CORRECT MOS: {correct_mos} months")
    print()
    print("-" * 50)
    print()
```

```python
# UPDATED: categorize_data_quality function for Cell 69
# Replace your existing function with this one

def categorize_data_quality(row):
    """
    Categorizes each item based on data availability and stock status.
    Uses the CORRECT months_of_supply from our unit conversion.
    """
    
    # FIRST: Check if we have consumption data
    if pd.isna(row.get('monthly_usage_base_units', None)) or row.get('monthly_usage_base_units', 0) == 0:
        return 'NEEDS_CLINICAL_REVIEW'
    
    # SECOND: Check for stockouts (using corrected base units)
    if row.get('stock_base_units', 0) == 0:
        return 'STOCKOUT (ORDER NOW)'
    
    # THIRD: Check using CORRECT months_of_supply
    mos = row.get('months_of_supply', None)
    
    if pd.isna(mos):
        return 'NEEDS_CLINICAL_REVIEW'
    elif mos < 1:
        return 'CRITICAL LOW (<1 month)'
    elif mos < 3:
        return 'LOW STOCK (1-3 months)'
    elif mos < 6:
        return 'HEALTHY (3-6 months)'
    elif mos < 12:
        return 'ADEQUATE (6-12 months)'
    else:
        return 'OVERSTOCK (>12 months)'
```

```python
# Apply updated categorization to your cleaned data
stock_clean_final['data_quality_v2'] = stock_clean_final.apply(categorize_data_quality, axis=1)

print("=== UPDATED DATA QUALITY DISTRIBUTION ===")
print()
print(stock_clean_final['data_quality_v2'].value_counts())
print()
print("Sample of categorized items:")
print(stock_clean_final[['item_name', 'stock_base_units', 'monthly_usage_base_units', 'months_of_supply', 'data_quality_v2']].head(15).to_string())
```
